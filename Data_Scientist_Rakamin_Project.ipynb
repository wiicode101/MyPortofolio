```python
import pandas as pd

# Gantilah 'nama_file.csv' dengan nama file yang sesuai
data_pelanggan = pd.read_csv('/content/drive/MyDrive/Portofolio/Data Analyst/Marketing_Campaign_Data_A-21492758-f47a-4913-971e-68c9a721311a.csv')
data_pelanggan.head()
```

```python
# Ganti kolom 'Unnamed: 0' dengan 'SK_ID_CURR'
data_pelanggan.rename(columns={'Unnamed: 0': 'SK_ID_CURR'}, inplace=True)

data_pelanggan.head()
```

```python
data_pelanggan.info()
```

- "Late Ten" : Age<20
- "Twenty" : 20 <= Age<30
- "Thirty" : 30 <= Age<40
- "Fourty" : 40 <= Age<50
- "Fifty" : 50 <= Age<60
- "Above_Sixty" : 60 <= Age
Sebagai catatan, tim marketing hanya membutuhkan data pelanggan yang tidak memiliki kesulitan
pembayaran (karena ini terkait dengan kemampuan mereka dalam membeli produk) dan tinggal di wilayah
yang diberi rating 3 (REGION_RATING_CLIENT_W_CITY) serta melakukan transaksi pada tanggal 1 Januari
2017 - 31 Desember 2018.

```python
data_pelanggan['DATE_BIRTH'] = pd.to_datetime(data_pelanggan['DATE_BIRTH'], format='%d-%m-%Y')
data_pelanggan['APPL_DATETIME'] = pd.to_datetime(data_pelanggan['APPL_DATETIME'], format='%Y-%m-%d')
```

```python
filtered_data = data_pelanggan.loc[
    (data_pelanggan['TARGET'] == 0) &
    (data_pelanggan['REGION_RATING_CLIENT_W_CITY'] == 3) &
    (data_pelanggan['APPL_DATETIME'] >= '2017-01-01') &
    (data_pelanggan['APPL_DATETIME'] <= '2018-12-31')
].copy()
filtered_data.head()
```

```python
# Fungsi untuk mengelompokkan umur
def categorize_age(age):
    if age < 20:
        return 'Late_Ten'
    elif age < 30:
        return 'Twenty'
    elif age < 40:
        return 'Thirty'
    elif age < 50:
        return 'Fourty'
    elif age < 60:
        return 'Fifty'
    else:
        return 'Above_Sixty'
```

```python
# Terapkan fungsi kategorisasi umur pada data
filtered_data['Age_Group'] = (filtered_data['APPL_DATETIME'] - filtered_data['DATE_BIRTH']).dt.days // 365
filtered_data['Age_Group'] = filtered_data['Age_Group'].apply(categorize_age)

# Urutkan hasil sesuai dengan urutan umur
result = filtered_data['Age_Group'].value_counts().reindex(['Late_Ten', 'Twenty', 'Thirty', 'Fourty', 'Fifty', 'Above_Sixty'])

# Tampilkan hasil
print(result)
```

```python
import pandas as pd

# Gantilah 'nama_file.csv' dengan nama file yang sesuai
data1 = pd.read_csv('/content/drive/MyDrive/Portofolio/Data Analyst/hci_previous_application_0-9cc2b69c-f469-4d66-bfdf-48b23f2c006a.csv')
# # Ganti kolom 'Unnamed: 0' dengan 'SK_ID_CURR'
# df.rename(columns={'Unnamed: 0': 'SK_ID_CURR'}, inplace=True)
data1.head()
```

```python
# Gantilah 'nama_file.csv' dengan nama file yang sesuai
data2 = pd.read_csv('/content/drive/MyDrive/Portofolio/Data Analyst/hci_application_0-2f6a4257-a1e4-4f59-88a4-d425c8ae1eaf.csv')
# # Ganti kolom 'Unnamed: 0' dengan 'SK_ID_CURR'
# df.rename(columns={'Unnamed: 0': 'SK_ID_CURR'}, inplace=True)
data2.head()
```

```python
# Menggabungkan data berdasarkan kolom SK_ID_CURR
merged_data = pd.merge(data2, data1, on='SK_ID_CURR', how='inner')
merged_data.head()
```

```python
# Menyaring data sesuai kondisi yang diberikan
filtered_data = merged_data[
    (merged_data['CODE_GENDER'] == 'F') &
    (merged_data['CNT_CHILDREN'] > 0) &
    (merged_data['TARGET'] == 0) &
    (merged_data['NAME_CONTRACT_STATUS'] == 'Approved')
]
filtered_data.head()
```

```python
# Kondisi 1: Banyaknya pengajuan kredit yang disetujui dari pelanggan wanita yang memiliki anak dan tidak memiliki masalah/kesulitan dalam proses pembayaran (TARGET = 0)
result_condition_1 = len(filtered_data)

# Kondisi 2: Tipe kontrak (NAME_CONTRACT_TYPE) yang memiliki pengajuan kredit yang paling banyak disetujui dari pelanggan wanita yang memiliki anak dan tidak memiliki masalah/kesulitan dalam proses pembayaran (TARGET = 0)
most_approved_contract_type = filtered_data['NAME_CONTRACT_TYPE'].value_counts().idxmax()

# Kondisi 3: Banyaknya pengajuan kredit yang disetujui dari pelanggan wanita yang memiliki anak dan tidak memiliki masalah/kesulitan dalam proses pembayaran (TARGET = 0) untuk tipe kontrak Consumer Loans
result_condition_3 = len(filtered_data[filtered_data['NAME_CONTRACT_TYPE'] == 'Consumer loans'])

# Kondisi 4: Tipe kontrak (NAME_CONTRACT_TYPE) yang memiliki pengajuan kredit yang paling banyak disetujui kedua dari pelanggan wanita yang memiliki anak dan tidak memiliki masalah/kesulitan dalam proses pembayaran (TARGET = 0)
second_most_approved_contract_type = filtered_data['NAME_CONTRACT_TYPE'].value_counts().index[1]

# Kondisi 5: Banyaknya pengajuan kredit yang disetujui dari pelanggan wanita yang memiliki anak dan tidak memiliki masalah/kesulitan dalam proses pembayaran (TARGET = 0) untuk tipe kontrak Cash Loans
result_condition_5 = len(filtered_data[filtered_data['NAME_CONTRACT_TYPE'] == 'Cash loans'])

# Menampilkan hasil
print(f"Banyaknya pengajuan kredit yang disetujui: {result_condition_1}")
print(f"Tipe kontrak paling banyak disetujui: {most_approved_contract_type}")
print(f"Banyaknya pengajuan kredit Consumer Loans yang disetujui: {result_condition_3}")
print(f"Tipe kontrak kedua paling banyak disetujui: {second_most_approved_contract_type}")
print(f"Banyaknya pengajuan kredit Cash Loans yang disetujui: {result_condition_5}")
```
